This simple code generates random circles on a canvas to create an abstract art effect


```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Hide the axes
ax.set_axis_off()

# Generate random circles
for _ in range(20):  # You can change this to generate more or fewer circles
    # Random position for the center of the circle
    x_center = np.random.uniform(0, 1)
    y_center = np.random.uniform(0, 1)

    # Random radius
    radius = np.random.uniform(0.05, 0.2)

    # Random color
    color = np.random.rand(3,)  # Random RGB color

    # Draw the circle
    circle = plt.Circle((x_center, y_center), radius, color=color, alpha=0.6)
    ax.add_patch(circle)

# Show the artwork
plt.show()
```
